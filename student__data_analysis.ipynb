```python

import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Study Hours": [1, 2, 3, 4, 5, 6, 7],
    "Scores": [20, 30, 50, 60, 70, 85, 95]
}

df = pd.DataFrame(data)

df
```

```python
print("Insight: There is a positive relationship between study hours and scores. Students who study more tend to perform better.")
```

```python
df.describe()
```

```python

plt.plot(df["Study Hours"], df["Scores"])
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Study Hours vs Scores")
plt.show()
```

```python
print("Insight: As study hours increase, student scores also increase.")
```

```python
print("Insight: There is a positive relationship between study hours and scores. Students who study more tend to perform better.")
```

```python

plt.plot(df["Study Hours"], df["Scores"], marker='o')
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Student Performance Analysis")
plt.grid()
plt.show()
```
