This code performs the following steps:

1.  **Imports Libraries**: It imports `LinearRegression` from `sklearn.linear_model` for building the regression model and `numpy` for numerical operations.
2.  **Prepares Data**: It creates two NumPy arrays: `hours` (input feature) and `scores` (target variable). `hours` is reshaped to be a 2D array, which is required by scikit-learn models.
3.  **Creates and Trains Model**: A `LinearRegression` model is initialized and then trained (fitted) using the `hours` and `scores` data.
4.  **Makes Prediction**: It predicts the score for a new input, 6 hours of study, using the trained model.
5.  **Prints Result**: Finally, it prints the predicted score.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Data: study hours vs scores
hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
scores = np.array([20, 40, 50, 70, 90])

# Create model
model = LinearRegression()
model.fit(hours, scores)

# Predict
predicted_score = model.predict([[6]])

print("Predicted score for 6 hours study:", predicted_score)
```
